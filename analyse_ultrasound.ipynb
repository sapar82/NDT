```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import signal
from scipy.fft import fft, fftfreq, fftshift
```

```python
measures = 'measures2/moyens_cylindres_trains_first_20230928_101803_Traces.csv'
# header is every line with % except the last one
first_measure = pd.read_csv(measures, header=9)
fig, ax = plt.subplots()
print(first_measure.columns)
t = np.array(first_measure[first_measure.columns[0]])
y = np.array(first_measure[first_measure.columns[1]])
ax.plot(t, y)
index = [00 + i*123 for i in range(8)]
#add vline at each index
for i in index:
    ax.axvline(t[i], color='r')
```

```python
fig, ax = plt.subplots(1, 6)
for i in range(6):
    ax[i].plot(fftshift(np.imag(fft(y[index[i]:index[i+1]]))))
```
